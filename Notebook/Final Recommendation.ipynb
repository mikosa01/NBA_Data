```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Load the feature importances dataset
feat = pd.read_csv('gb_feat.csv')
feat.head()
```

```python
#Show the list of column names  
feat.columns
```

```python
#Drop unwanted columns 
feat.drop('Unnamed: 0', 1, inplace = True)
```

```python
#show the top five rows of the feature table
feat.head()
```

```python
#Load the players stats csv
stats = pd.read_csv('part-00000-a839c6d4-e4d8-46ff-aedb-5aeb877960f5-c000.csv')
```

```python
#Showing the five columns
stats.head()
```

```python
#List the all the values in your Feature column
feat_col = [x for x in feat['Feature']]
feat_col
```

```python
#Creating a subset of the stats table using the feat_col
new_df = stats[feat_col]
```

```python
#Calculating the average of new_df columns. 
feat['avg_game'] = [np.round(new_df[x].mean(), 0) for x in new_df.columns]
```

```python
feat
```

```python
#Calculate the feature in percentage 
feat['Impo_pct'] = feat['Importance'] *100
```

```python
feat
```

```python
#Calculate the average of the feature importance
feat['avg_impo'] = feat['Impo_pct'] /2
```

```python
#Calculate the least points each players is expected to get at each game.
feat['lowest_pt'] = np.round((feat['avg_game'] * feat['avg_impo']) / feat['Impo_pct'])
```

```python
feat
```

```python
#Creating a table of just Feature and lowest point table for comparism
new_feat = feat[['Feature', 'lowest_pt']]
```

```python
new_df.head()
```

```python
#Selecting a players game stats from the stats table
play_stats = new_df.loc[4].to_frame().reset_index()
play_stats= play_stats.rename(columns = {4:'stat'})
```

```python
play_stats
```

```python
#Merging with the least point table
data = new_feat.merge(play_stats['stat'], left_on = 'Feature', right_on= play_stats['index'])
```

```python
#Creating a recommendation for the player
for index, row in data.iterrows():
    if row['lowest_pt'] > row['stat']:
        print('Improve the your ', format(row['Feature']))
```
